```python
import csv
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from keras.utils import np_utils
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.models import Model
from keras.layers import Dense, Dropout, Activation, Flatten, Input
from keras.layers import Convolution2D, MaxPooling2D, AveragePooling2D
from keras.optimizers import Adam
from keras import backend as K
import glob
from PIL import Image
import json
import h5py
```

#### 使用する画像：cat, dog, frog

```python
frog_list = []
dog_list = []
cat_list = []
```

```python
with open('trainLabels.csv', 'r') as f:
    reader = csv.reader(f)
    header = next(reader)
    for row in reader:
        image_name = './train/' + str(row[0]) + '.png'
        if row[1] == 'frog':
            frog_list.append(image_name)
        elif row[1] == 'dog':
            dog_list.append(image_name)
        elif row[1] == 'cat':
            cat_list.append(image_name)
```

```python
image = np.empty((0,3,32,32), dtype=np.uint8)
result = np.empty((0,3), dtype=np.uint8)

for i in frog_list:
    im_reading = np.array( Image.open(i).resize((32,32)))
    im_reading = im_reading.transpose(2,0,1)
    image = np.append(image, [im_reading], axis=0)
    result = np.append(result, np.array([[1,0, 0]]), axis=0)

for i in dog_list:
    im_reading = np.array( Image.open(i).resize((32,32)))
    im_reading = im_reading.transpose(2,0,1)
    image = np.append(image, [im_reading], axis=0)
    result = np.append(result, np.array([[0,1, 0]]), axis=0)
    
for i in cat_list:
    im_reading = np.array( Image.open(i).resize((32,32)))
    im_reading = im_reading.transpose(2,0,1)
    image = np.append(image, [im_reading], axis=0)
    result = np.append(result, np.array([[0,0, 1]]), axis=0)
```

```python
number = 1
print(image.shape)
plt.imshow( image[number].transpose(1,2,0) )
print(result[number])
```

```python
# http://stackoverflow.com/questions/3674409/numpy-how-to-split-partition-a-dataset-array-into-training-and-test-datasets
from sklearn.cross_validation import train_test_split
data_train, data_test, labels_train, labels_test = train_test_split(image, result, test_size=0.10, random_state=10)
```

### Sequential Model

```python
model = Sequential()
model.add(Convolution2D(16, 3, 3, border_mode="same", activation="relu" ,input_shape=(3, 32, 32) ))
model.add(Convolution2D(8, 3, 3, border_mode="same", activation="relu"))
model.add(Convolution2D(4, 3, 3, border_mode="same", activation="relu"))
model.add(Dropout(0.5))

model.add(Flatten())
model.add(Dense(512))
model.add(Activation("relu"))
model.add(Dense(10))
model.add(Activation("relu"))
model.add(Dense(3))
model.add(Activation("sigmoid"))
model.compile(loss='binary_crossentropy', optimizer="adadelta", metrics=['accuracy'])

json_string = model.to_json()
open(os.path.join('./','cnn_model.json'), 'w').write(json_string)
```

```python
hist = model.fit(data_train, labels_train, nb_epoch=20, batch_size=32, validation_data=(data_test, labels_test))
```

```python
loss = hist.history['loss']
val_loss = hist.history['val_loss']
```

```python
nb_epoch = len(loss)
plt.plot(range(nb_epoch), loss, marker='.', label='loss')
plt.plot(range(nb_epoch), val_loss, marker='.', label='val_loss')
plt.legend(loc='best', fontsize=10)
plt.grid()
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

### Functional Model

```python
image_input = Input(shape=(3, 32, 32))

x1 = Convolution2D(16, 3, 3, border_mode="same", activation="relu")(image_input)
x2 = Convolution2D(8, 3, 3, border_mode="same", activation="relu")(x1)
x3 = Convolution2D(4, 3, 3, border_mode="same", activation="relu")(x2)
x_dropout = Dropout(0.5)(x3)
x_flatten = Flatten()(x_dropout)
x_dense1 = Dense(512)(x_flatten)
x_activation1 = Activation("relu")(x_dense1)
x_dense2 = Dense(10)(x_activation1)
x_activation2 = Activation("relu")(x_dense2)
x_dense3 = Dense(3)(x_activation2)
x_activation3 = Activation("sigmoid")(x_dense3)

functional_model = Model(input=image_input, output=x_activation3)

functional_model.compile(loss='binary_crossentropy', optimizer="adadelta", metrics=['accuracy'])

json_string = model.to_json()
open(os.path.join('./','cnn_functional_model.json'), 'w').write(json_string)
```
